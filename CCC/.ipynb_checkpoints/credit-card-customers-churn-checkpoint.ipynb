```python
#'Attrition_Flag'
churn_count = customers['Attrition_Flag'].value_counts().reset_index()
churn_count.columns = ['Customer status','Count']
churn_count['Percentage'] = churn_count['Count'].apply(lambda x: round(x/churn_count['Count'].sum()*100),0)
churn_count
```
